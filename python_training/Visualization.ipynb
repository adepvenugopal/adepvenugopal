# **Import Packages**

```python
#Import pandas package
import pandas as pd
```

# **Import Data**

```python
#Read nyc_weather.csv into df
df = pd.read_csv('nyc_weather.csv')
```

```python
#Find the count of each category
df.Events.value_counts()
```

```python
#Bar plot the count of each category
df.Events.value_counts().plot.bar()
```

```python
#Bar plot the count of each category (Sort in ascending order)
df.Events.value_counts().sort_values().plot.bar()
```

```python
#Bar plot the count of each category (Sort in descending order)
df.Events.value_counts().sort_values(ascending=False).plot.bar()
```

```python
#Line plot the count of each category
df.Events.value_counts().plot.line()
```

```python
#Bar plot the count of each category (Sort in ascending order)
df.Events.value_counts().sort_values().plot.bar()
```

```python
#Bar plot the count of each category (Sort in descending order)
df.Events.value_counts().sort_values(ascending=False).plot.bar()
```

```python
#Line plot the count of each category
df.Events.value_counts().plot.line()
```

```python
#Area plot the count of each category
df.Events.value_counts().plot.area()
```
